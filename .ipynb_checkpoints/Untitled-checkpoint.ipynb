```python
import os
from tkinter import *
from tkinter import Tk
from tkinter import filedialog
import pygame
import numpy as np
import io
```

```python
class MusicPlayer():
    def __init__(self,root):
        self.root = root
        # Title of the window
        self.root.title("MusicPlayer")
        # Window Geometry
        self.root.geometry("1000x200+200+200")
        # Initiating Pygame
        pygame.init()
        # Initiating Pygame Mixer
        pygame.mixer.init()
        # Declaring track Variable
        self.track = StringVar()
        # Declaring Status Variable
        self.status = StringVar()

        # Creating the Track Frames for Song label & status label
        trackframe = LabelFrame(self.root,text="Song Track",font=("times new roman",15,"bold"),bg="Navyblue",fg="white",bd=5,relief=GROOVE)
        trackframe.place(x=0,y=0,width=600,height=100)
        # Inserting Song Track Label
        songtrack = Label(trackframe,textvariable=self.track,width=20,font=("times new roman",24,"bold"),bg="Orange",fg="gold").grid(row=0,column=0,padx=10,pady=5)
        # Inserting Status Label
        trackstatus = Label(trackframe,textvariable=self.status,font=("times new roman",24,"bold"),bg="orange",fg="gold").grid(row=0,column=1,padx=10,pady=5)

        # Creating Button Frame
        buttonframe = LabelFrame(self.root,text="Control Panel",font=("times new roman",15,"bold"),bg="grey",fg="white",bd=5,relief=GROOVE)
        buttonframe.place(x=0,y=100,width=600,height=100)
        # Inserting Play Button
        playbtn = Button(buttonframe,text="PLAYSONG",command=self.playsong,width=10,height=1,font=("times new roman",16,"bold"),fg="navyblue",bg="pink").grid(row=0,column=0,padx=10,pady=5)
        # Inserting Pause Button
        playbtn = Button(buttonframe,text="PAUSE",command=self.pausesong,width=8,height=1,font=("times new roman",16,"bold"),fg="navyblue",bg="pink").grid(row=0,column=1,padx=10,pady=5)
        # Inserting Unpause Button
        playbtn = Button(buttonframe,text="UNPAUSE",command=self.unpausesong,width=10,height=1,font=("times new roman",16,"bold"),fg="navyblue",bg="pink").grid(row=0,column=2,padx=10,pady=5)
        # Inserting Stop Button
        playbtn = Button(buttonframe,text="STOPSONG",command=self.stopsong,width=10,height=1,font=("times new roman",16,"bold"),fg="navyblue",bg="pink").grid(row=0,column=3,padx=10,pady=5)

        # Creating Playlist Frame
        songsframe = LabelFrame(self.root,text="Song Playlist",font=("times new roman",15,"bold"),bg="grey",fg="white",bd=5,relief=GROOVE)
        songsframe.place(x=600,y=0,width=400,height=200)
        # Inserting scrollbar
        scrol_y = Scrollbar(songsframe,orient=VERTICAL)
        # Inserting Playlist listbox
        self.playlist = Listbox(songsframe,yscrollcommand=scrol_y.set,selectbackground="gold",selectmode=SINGLE,font=("times new roman",12,"bold"),bg="silver",fg="navyblue",bd=5,relief=GROOVE)
        # Applying Scrollbar to listbox
        scrol_y.pack(side=RIGHT,fill=Y)
        scrol_y.config(command=self.playlist.yview)
        self.playlist.pack(fill=BOTH)
        # Changing Directory for fetching Songs
        os.chdir("D:/lpnu/магістратура/диплом/musdb18_wav/train")
        # Fetching Songs
        songtracks = os.listdir()
        # Inserting Songs into Playlist
        for track in songtracks:
            self.playlist.insert(END,track)
            
    def playsong(self):
        # Displaying Selected Song title
        self.track.set(self.playlist.get(ACTIVE))
        # Displaying Status
        self.status.set("-Playing")
        # Loading Selected Song
        pygame.mixer.music.load(self.playlist.get(ACTIVE))
        # Playing Selected Song
        pygame.mixer.music.play()
            
    def stopsong(self):
        # Displaying Status
        self.status.set("-Stopped")
        # Stopped Song
        pygame.mixer.music.stop()
            
    def pausesong(self):
        # Displaying Status
        self.status.set("-Paused")
        # Paused Song
        pygame.mixer.music.pause()
            
    def unpausesong(self):
        # It will Display the  Status
        self.status.set("-Playing")
        # Playing back Song
        pygame.mixer.music.unpause()
```

```python
root = Tk() # In order to create an empty window
# Passing Root to MusicPlayer Class
MusicPlayer(root)
```

```python
root.mainloop()
```

```python
import sys
from PyQt5.QtWidgets import QApplication, QMainWindow, QWidget, QVBoxLayout, QHBoxLayout, QPushButton, QSlider, QFileDialog
from PyQt5.QtCore import Qt, QTimer
from pydub import AudioSegment
from pydub.playback import play
import librosa
import numpy as np
```

```python
class AudioPlayer(QMainWindow):
    def __init__(self):
        super().__init__()

        self.channels = 4
        self.sample_width = 2
        self.frame_rate = 44100

        self.audio_file = None
        self.channel_data = [None] * self.channels
        self.playing = False
        self.muted = [False] * self.channels
        self.current_position = 0

        self.initUI()
        self.timer = QTimer(self)
        self.timer.timeout.connect(self.update_progress)

    def initUI(self):
        central_widget = QWidget(self)
        self.setCentralWidget(central_widget)
        layout = QVBoxLayout(central_widget)

        open_button = QPushButton('Open WAV File', self)
        open_button.clicked.connect(self.open_wav)
        layout.addWidget(open_button)

        control_layout = QHBoxLayout()

        play_button = QPushButton('Play', self)
        play_button.clicked.connect(self.play)
        control_layout.addWidget(play_button)

        stop_button = QPushButton('Stop', self)
        stop_button.clicked.connect(self.stop)
        control_layout.addWidget(stop_button)

        move_to_start_button = QPushButton('Move to Start', self)
        move_to_start_button.clicked.connect(self.move_to_start)
        control_layout.addWidget(move_to_start_button)

        layout.addLayout(control_layout)

        channel_layout = QVBoxLayout()
        self.sliders = []

        for i in range(self.channels):
            channel_layout.addLayout(self.create_channel_controls(i))

        layout.addLayout(channel_layout)

        self.update_ui()

    def create_channel_controls(self, channel_index):
        channel_controls = QHBoxLayout()

        mute_button = QPushButton(f'Mute Channel {channel_index + 1}', self)
        mute_button.clicked.connect(lambda ch=channel_index: self.toggle_mute(ch))
        channel_controls.addWidget(mute_button)

        slider = QSlider(Qt.Horizontal)
        slider.valueChanged.connect(lambda value, ch=channel_index: self.update_position(value, ch))
        self.sliders.append(slider)
        channel_controls.addWidget(slider)

        return channel_controls

    def open_wav(self):
        options = QFileDialog.Options()
        options |= QFileDialog.ReadOnly
        file_path, _ = QFileDialog.getOpenFileName(self, "Open WAV File", "", "WAV Files (*.wav);;All Files (*)", options=options)

        if file_path:
            audio_data, sample_rate = librosa.load(file_path, sr=None, mono=False)

            self.frame_rate = sample_rate
            self.sample_width = 2  # You can set this as needed

            for i in range(1, self.channels + 1):
                if i <= len(audio_data):
                    self.channel_data[i] = (audio_data[i] * (2 ** 15)).astype(np.int16)

            self.current_position = 0
            self.update_ui()

    def play(self):
        if any(self.channel_data[i] is not None for i in range(1, self.channels)):
            if not self.playing:
                audio_to_play = AudioSegment.silent(duration=1, frame_rate=self.frame_rate)
                for i in range(1, self.channels):
                    if self.channel_data[i] is not None:
                        audio_to_play = audio_to_play.overlay(
                            AudioSegment.from_raw(
                                self.channel_data[i][self.current_position:],
                                frame_rate=self.frame_rate,
                                sample_width=self.sample_width,
                                channels=1
                            )
                        )
                play(audio_to_play)
                self.playing = True

    def stop(self):
        self.playing = False

    def move_to_start(self):
        self.current_position = 0
        self.update_ui()

    def toggle_mute(self, channel_index):
        self.muted[channel_index] = not self.muted[channel_index]
        self.update_ui()

    def update_position(self, value, channel_index):
        if not self.playing:
            self.current_position = value
            self.update_ui()

    def update_progress(self):
        if self.playing:
            self.playing = False  # Stop playback at the next update
            self.update_ui()

    def update_ui(self):
        if any(self.channel_data[i] is not None for i in range(1, self.channels)):
            for channel_index in range(1, self.channels):
                self.sliders[channel_index].setRange(0, len(self.channel_data[1]))
                self.sliders[channel_index].setValue(self.current_position)
                self.sliders[channel_index].setEnabled(not self.muted[channel_index])
```

```python
def main():
    app = QApplication(sys.argv)
    player = AudioPlayer()
    player.show()
    sys.exit(app.exec_())
```

```python
if __name__ == '__main__':
    main()
```

```python
import sys
import os
from PyQt5.QtWidgets import QApplication, QMainWindow, QFileDialog, QPushButton, QSlider, QLabel
from PyQt5.QtMultimedia import QMediaContent, QMediaPlayer
from PyQt5.QtCore import Qt, QUrl  # Added import statement
```

```python
class AudioPlayer(QMainWindow):
    def __init__(self):
        super().__init__()

        self.player = QMediaPlayer(None, QMediaPlayer.StreamPlayback)
        self.init_ui()

    def init_ui(self):
        self.setWindowTitle("Audio Player")
        self.setGeometry(100, 100, 400, 200)

        self.play_button = QPushButton("Play", self)
        self.play_button.setGeometry(50, 20, 100, 30)
        self.play_button.clicked.connect(self.play_audio)

        self.pause_button = QPushButton("Pause", self)
        self.pause_button.setGeometry(160, 20, 100, 30)
        self.pause_button.clicked.connect(self.pause_audio)

        self.mute_button = QPushButton("Mute", self)
        self.mute_button.setGeometry(270, 20, 100, 30)
        self.mute_button.clicked.connect(self.toggle_mute)

        self.progress_slider = QSlider(Qt.Horizontal, self)
        self.progress_slider.setGeometry(50, 70, 320, 30)
        self.progress_slider.sliderMoved.connect(self.set_position)

        self.progress_label = QLabel("00:00 / 00:00", self)
        self.progress_label.setGeometry(50, 110, 320, 30)

        self.load_button = QPushButton("Load Audio", self)
        self.load_button.setGeometry(50, 160, 100, 30)
        self.load_button.clicked.connect(self.load_audio)

    def on_media_status_changed(self, status):
        if status == QMediaStatus.LoadedMedia:
            self.update_duration(self.player.duration())
        
    def load_audio(self):
        options = QFileDialog.Options()
        options |= QFileDialog.ReadOnly
        audio_file, _ = QFileDialog.getOpenFileName(self, "Open Audio File", "", "Audio Files (*.mp4 *.mp3 *.wav);;All Files (*)", options=options)

        if audio_file:
            print(self.player.isAudioAvailable())
            self.player.setMedia(QMediaContent(QUrl.fromLocalFile(audio_file)))
            print(self.player.isAudioAvailable())

    def play_audio(self):
        print(self.player.mediaStatus())
        print(QMediaPlayer.LoadedMedia)
        print(self.player.duration())
        if self.player.mediaStatus() == QMediaPlayer.StoppedState:
            self.player.play()
            self.player.durationChanged.connect(self.update_duration)
            self.player.positionChanged.connect(self.update_position)

    def pause_audio(self):
        self.player.pause()

    def toggle_mute(self):
        self.player.setMuted(not self.player.isMuted())

    def update_duration(self, duration):
        minutes = duration // 60000
        seconds = (duration // 1000) % 60
        self.progress_slider.setRange(0, duration)
        self.progress_label.setText(f"00:00 / {minutes:02}:{seconds:02}")

    def update_position(self, position):
        minutes = position // 60000
        seconds = (position // 1000) % 60
        duration = self.player.duration()
        self.progress_slider.setValue(position)
        self.progress_label.setText(f"{minutes:02}:{seconds:02} / {duration // 60000:02}:{(duration // 1000) % 60:02}")

    def set_position(self, position):
        self.player.setPosition(position)
```

```python
if __name__ == "__main__":
    app = QApplication([])
    player = AudioPlayer()
    player.show()
    app.exec_()
```
